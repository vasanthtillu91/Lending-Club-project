```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import linear_model
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn import neighbors
import graphviz
```

```python
LC_data=pd.read_csv("H:\Fall 2017\Applied Machine Learning\Projects\Individual_Project\Individual Project files\FINAL_DATA_CONSIDER\CSV\combi_final.csv")
```

```python
LC_data
```

```python
pd.isnull(LC_data).any()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
LC_data2=LC_data.apply(LabelEncoder().fit_transform)
```

```python
LC_data2
```

```python
LC_data_features=LC_data2.drop('loan_status(Target)',1)
```

```python
LC_data_features
```

```python
LC_target=LC_data2['loan_status(Target)']
label_name = ["Good", "Bad"]
```

```python
LC_target
```

```python
#75-25 split:

X_train, X_test, Y_train, Y_test = train_test_split(LC_data_features, LC_target, test_size=0.25, random_state=0)
```

```python
# 80-20 split on 75% data

X_train1, X_test1, Y_train1, Y_test1 = train_test_split(X_train, Y_train, test_size=0.2, random_state=0)
```

```python
regr=linear_model.LogisticRegression()
regr.fit(X_train1,Y_train1)
```

```python
print('Coefficients:', regr.coef_)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score
```

```python
## Cross Validation on 80% of Initial 75% data

accuracy_lg=cross_val_score(regr, X_train1, Y_train1, cv=10)
precision_lg = cross_val_score(regr, X_train1, Y_train1, cv=10, scoring='precision')
recall_lg = cross_val_score(regr, X_train1, Y_train1, cv=10, scoring='recall')
f1_lg = cross_val_score(regr, X_train1, Y_train1, cv=10, scoring='f1')
```

```python
print("Accuracy_lg:",accuracy_lg)
print("Precision_lg:",precision_lg)
print("Recall_lg:",recall_lg)
print("F1 score_lg:",np.mean(f1_lg), f1_lg)
```

```python
regr_score_train1=regr.score(X_train1,Y_train1)
regr_score_test1=regr.score(X_test1,Y_test1)
print('Training Score:', regr_score_train1)
print('Test score',regr_score_test1)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
```

```python
# Testing the model on 20% data of Initial 75% Data

test_prediction_regr = regr.predict(X_test1)
print ('Accuracy1_regr:', accuracy_score(Y_test1, test_prediction_regr))
print ('F1 score_regr:', f1_score(Y_test1, test_prediction_regr,average='weighted'))
print ('Recall_regr:', recall_score(Y_test1, test_prediction_regr,
                              average='weighted'))
print ('Precision_regr:', precision_score(Y_test1, test_prediction_regr,
                                    average='weighted'))
```

```python
# Testing the model on initial 25% data

test_prediction_regr = regr.predict(X_test)
print ('Accuracy1_regr:', accuracy_score(Y_test, test_prediction_regr))
print ('F1 score2_regr:', f1_score(Y_test, test_prediction_regr,average='weighted'))
print ('Recall2_regr:', recall_score(Y_test, test_prediction_regr,
                              average='weighted'))
print ('Precision2_regr:', precision_score(Y_test, test_prediction_regr,
                                    average='weighted'))
```
