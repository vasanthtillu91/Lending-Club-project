```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
loan_numeric_data=pd.read_csv("H:\Fall 2017\Applied Machine Learning\Projects\Individual_Project\Individual Project files\data_play\PCA_numeric_data\PCA_numeric_data.csv")
```

```python
loan_numeric_data
```

```python
#### Imputing missing values before feeding data for PCA
```

```python
df = pd.DataFrame(loan_numeric_data)
```

```python
df
```

```python
from sklearn.preprocessing import Imputer
imp = Imputer(missing_values='NaN', strategy='most_frequent', axis=0)
imp.fit(df)
Imputer(axis=0, copy=True, missing_values='NaN', strategy='most_frequent', verbose=0)
print(imp.transform(df))
```

```python
df2=imp.transform(df)
```

```python
### Now applying the PCA on the imputed dataset: Mod_loan_numeric_data
pca = PCA(n_components=10)
pca.fit(df2)
```

```python
print(pca.explained_variance_ratio_)
```

```python
pca.components_
```

```python
#pd.DataFrame(pca.transform(df2), columns=['PCA%i' % i for i in range(n_components)], index=df2.index)
```

```python
df2_PCA = pca.transform(df2)
df2_PCA
```

```python
PCA_loan_numeric_data=pd.DataFrame(df2_PCA,columns=('pca1','pca2','pca3','pca4','pca5','pca6','pca7','pca8','pca9','pca10'))
```

```python
PCA_loan_numeric_data
```

```python
#### The first 7 Columns in mutual combination, explain about 99.5% variance in whole data. So dropping last 3 columns
```

```python
EFF_PCA_FINAL=PCA_loan_numeric_data.drop(['pca8', 'pca9','pca10'],axis=1)
```
